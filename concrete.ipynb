### **Data Pre-processing**

```
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```
stdscaler = StandardScaler()
X_train = stdscaler.fit_transform(X_train)
X_test = stdscaler.fit_transform(X_test)
```

### **Building the model**

**Decision trees**


```
dt = DecisionTreeRegressor()
dt.fit(X_train,y_train)
pred_dt = dt.predict(X_test)

plt.figure(figsize=(10,7))
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Decision Tree Model")
plt.scatter(y_test,pred_dt, color = 'thistle')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k-')
plt.show()

print("RMSE \t\t MSE \t\t MAE \t\t R2")
print("""{:.2f} \t\t {:.2f} \t\t{:.2f} \t\t{:.2f}""".format(
            np.sqrt(mean_squared_error(y_test, pred_dt)),mean_squared_error(y_test, pred_dt),
            mean_absolute_error(y_test, pred_dt), r2_score(y_test, pred_dt)))

```

**Random Forest Regressor**

```
rf = RandomForestRegressor(n_estimators = 100)
rf.fit(X_train, y_train)
pred_rf = rf.predict(X_test)

plt.figure(figsize=(10,7))
plt.title("Random Forest Model")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.scatter(y_test, pred_rf, color = 'powderblue')
plt.show()

print("RMSE \t\t MSE \t\t MAE \t\t R2")
print("""{:.2f} \t\t {:.2f} \t\t{:.2f} \t\t{:.2f}""".format(
            np.sqrt(mean_squared_error(y_test, pred_rf)),mean_squared_error(y_test, pred_rf),
            mean_absolute_error(y_test, pred_rf), r2_score(y_test, pred_rf)))
```

**Comparing the models**

```
models = [dt,rf]
names = ["Decision Tree","Random Forest"]
rmse = []

for model in models : 
  rmse.append(np.sqrt(mean_squared_error(y_test,model.predict(X_test))))

x = np.arange(len(names))
width = 0.5
plt.figure(figsize=(7,5))
modelplot = plt.bar(x,rmse,width,color = ['thistle','powderblue'])
plt.xlabel("RMSE")
plt.title("RMSE with different models")
plt.ylabel("Models")
plt.xticks(x,names,rotation = False)
```

### **Conclusion**

Random Forest Regression provided the least RMSE. Hence, it is the best model for this predicting the compressive strength of concrete.

```
random_values = [[520,130,17.5,173,2.5,999,895,275]] #cement,slag,ash,water,plastic,coarse,fine,age
pred_rf = rf.predict(random_values)
pred_dt = dt.predict(random_values)
print("The estimated compressive strength of concrete using Random Forest is : ",pred_rf)
print("\nThe estimated compressive strength of concrete using Decision Tree is : ",pred_dt)
```
